```python
text = 'ピコ太郎さんはカナブンに角をつけてカブトムシとして売るバイトをしている'
```

```python
import MeCab

mecab = MeCab.Tagger("-Ochasen -d /usr/lib/mecab/dic/mecab-ipadic-neologd")
res = mecab.parse(text)
print(res)
```

```python
import spacy
from spacy import displacy

nlp = spacy.load("ja_ginza")


def render_parse(text, style="dep"):
    doc = nlp(text)
    displacy.render(
        doc, style=style, jupyter=True, options={"fine_grained": True, "compact": True}
    )
```

```python
render_parse(text)
```
